```python
url = "http://www.puneapmc.org/rates.aspx"
```

```python
html = helpers.scrape(url)
```

```python
from bs4 import BeautifulSoup
import requests
```

```python
soup = BeautifulSoup(html)
```

```python
h3_tag = soup.find('h3')
if h3_tag:
    # Find the first <a> tag in the next <ul> or <li> after the <h3> tag
    next_link = h3_tag.find_next('a')
    
    if next_link and 'href' in next_link.attrs:
        # Construct the full URL if it's relative
        link_url = next_link['href']
        if not link_url.startswith('http'):
            link_url = requests.compat.urljoin(url, link_url)
```

```python
link_url
```

```python
commodityHTML = helpers.scrape(link_url)
```

```python
soup = BeautifulSoup(commodityHTML)

# Find all tables in the HTML
tables = soup.find_all('table')

# Loop through each table
for index, table in enumerate(tables):
    
    # Initialize a list to store rows of the current table
    table_data = []
    
    # Find all rows in the table
    rows = table.find_all('tr')
    
    # Loop through rows starting from the second row (skipping the header)
    for row in rows[1:]:
        cells = row.find_all('td')
        
        # Ensure there are at least 6 cells in the row
        if len(cells) >= 6:
            # Extract and clean all data fields
            try:
                code_no = int(cells[0].text.strip())  # Convert to integer
            except ValueError:
                code_no = None  # Handle missing or invalid data gracefully
            
            commodity = cells[1].text.strip()
            quantity = cells[2].text.strip()
            
            try:
                income = int(cells[3].text.strip())  # Convert to integer
            except ValueError:
                income = None  # Handle missing or invalid data gracefully
            
            try:
                min_price = int(cells[4].text.strip().replace('Rs.', '').replace('/-', '').replace(',', '').strip())
            except ValueError:
                min_price = None  # Handle missing or invalid data gracefully

            try:
                max_price = int(cells[5].text.strip().replace('Rs.', '').replace('/-', '').replace(',', '').strip())
            except ValueError:
                max_price = None  # Handle missing or invalid data gracefully
            
            # Append the extracted data to the table's data list
            row_data = {
                            "code_no":code_no, 
                            "commodity":commodity, 
                            "quantity":quantity, 
                            "income":income, 
                            "min_price":min_price, 
                            "max_price":max_price
                            }
            table_data.append(row_data)

    # Print or process the table data
    # for row_data in table_data:
      #  print(row_data)

    print(table_data)
```
