```python
from selenium import webdriver
from selenium.webdriver import ChromeOptions
import time
```

```python
options = ChromeOptions()
```

```python
url = "https://agmarknet.gov.in/SearchCmmMkt.aspx?Tx_Commodity=4&Tx_State=MH&Tx_District=14&Tx_Market=172&DateFrom=01-Nov-2024&DateTo=13-Nov-2024&Fr_Date=01-Nov-2024&To_Date=13-Nov-2024&Tx_Trend=0&Tx_CommodityHead=Maize&Tx_StateHead=Maharashtra&Tx_DistrictHead=Pune&Tx_MarketHead=Pune"
```

```python
with webdriver.Chrome(options=options) as driver:
    driver.get(url)
    time.sleep(10)
    html = driver.page_source
```
